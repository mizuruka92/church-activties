```python
def likert_traffic_matrix(
    df,
    group_col,
    question_cols,
    group_values=None,
    likert_min=1,
    likert_max=5
):
    """
    Build traffic-light matrix (🟢🟡🔴) for Likert questions
    grouped by any categorical column.

    Returns:
        dot_matrix  : Question x Group with emoji status
        stats_table : Avg & P25 per Question x Group
    """

    import numpy as np
    import pandas as pd

    def extract_likert(series):
        return (
            series.astype(str)
            .str.strip()
            .str[0]
            .apply(pd.to_numeric, errors="coerce")
        )

    def traffic_light(avg, p25):
        if avg >= 4.0:
            if p25 < 3.0:
                return "🟡"
            return "🟢"
        elif avg >= 3.3:
            if p25 < 2.5:
                return "🔴"
            return "🟡"
        else:
            return "🔴"

    if group_values is None:
        group_values = (
            df[group_col]
            .dropna()
            .astype(str)
            .unique()
            .tolist()
        )

    results = []

    for question in question_cols:
        for group in group_values:
            subset = df[df[group_col] == group][question]
            scores = extract_likert(subset).dropna()

            if len(scores) == 0:
                avg = np.nan
                p25 = np.nan
                status = "⚪"
            else:
                avg = scores.mean()
                p25 = scores.quantile(0.25)
                status = traffic_light(avg, p25)

            results.append({
                "Question": question,
                "Group": group,
                "Avg": round(avg, 2) if not np.isnan(avg) else np.nan,
                "P25": round(p25, 2) if not np.isnan(p25) else np.nan,
                "Status": status
            })

    df_results = pd.DataFrame(results)

    dot_matrix = (
        df_results
        .pivot(index="Question", columns="Group", values="Status")
        .reindex(columns=group_values)
    )

    stats_table = (
        df_results
        .pivot_table(
            index="Question",
            columns="Group",
            values=["Avg", "P25"]
        )
    )

    return dot_matrix, stats_table
```

```python
placeholder = ['Saya adalah seorang', 'Status pernikahan saya', 'Pendidikan terakhir saya yang sudah saya selesaikan']
df = data[['Status pernikahan saya']+myquestions]
dot_wed, stats_wed = likert_traffic_matrix(
    df=df,
    group_col="Status pernikahan saya",
    question_cols=myquestions
)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pptx import Presentation
from pptx.util import Inches
import numpy as np
import re
import os

# ================= CONFIG =================

OUTPUT_DIR = "charts_3d_normalized"
PPT_PATH = "Survey_Report_3D_Normalized.pptx"

LIKERT_VALUES = [0, 1, 2, 3, 4, 5]

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ================= LOAD DATA =================


# ================= AGE GROUP =================

AGE_COLUMN = "Usia saya"

AGE_GROUPS = ["<17", "17–34", "35–65", ">65"]

AGE_COLORS = {
    "<17": "#4C72B0",     # blue
    "17–34": "#55A868",   # green
    "35–65": "#C44E52",   # red
    ">65": "#8172B2"      # purple
}

AGE_MAP = {
    "dibawah 17 Tahun": "<17",
    "17 ~ 24 Tahun": "17–34",
    "25 ~ 34 Tahun": "17–34",
    "35 ~ 49 Tahun": "35–65",
    "50 ~ 65 Tahun": "35–65",
    "Diatas 65 Tahun": ">65"
}
df["Age_Group"] = df[AGE_COLUMN].map(AGE_MAP)

# ================= HELPERS =================
def split_question(col):
    m = re.match(r"(.*?)\s*\[(.*?)\]", col)
    return (m.group(1).strip(), m.group(2).strip()) if m else (col, col)

def extract_likert(series):
    return (
        series.astype(str)
        .str.strip()
        .str[0]
        .apply(pd.to_numeric, errors="coerce")
    )
Y_SPACING = 1.5
# ================= PPT =================
prs = Presentation()
slide_layout = prs.slide_layouts[5]  # title only

# ================= PROCESS =================
for col in df.columns:
    if col in [AGE_COLUMN, "Age_Group"]:
        continue

    group_title, detail_title = split_question(col)

    # ---- RAW COUNTS ----
    raw = {ag: {lv: 0 for lv in LIKERT_VALUES} for ag in AGE_GROUPS}

    for ag in AGE_GROUPS:
        values = extract_likert(df[df["Age_Group"] == ag][col]).dropna()
        for lv in LIKERT_VALUES:
            raw[ag][lv] = (values == lv).sum()

    # ---- NORMALIZE ----
    norm = {}
    for ag in AGE_GROUPS:
        total = sum(raw[ag].values())
        norm[ag] = {
            lv: (raw[ag][lv] / total * 100) if total > 0 else 0
            for lv in LIKERT_VALUES
        }

    # ================= 3D BAR =================
    fig = plt.figure(figsize=(9, 6))
    ax = fig.add_subplot(111, projection="3d")
    ax.view_init(elev=5, azim=-80)

    xpos, ypos, zpos, dz, colors = [], [], [], [], []

    for y, ag in enumerate(AGE_GROUPS):
        for lv in LIKERT_VALUES:
            xpos.append(lv)
            ypos.append(y * Y_SPACING)
            zpos.append(0)
            dz.append(norm[ag][lv])
            colors.append(AGE_COLORS.get(ag, "#999999"))

    ax.bar3d(
        xpos, ypos, zpos,
        0.5, 0.5, dz,
        color=colors,
        shade=True
    )

    ax.set_xlabel("Skala Likert (0–5)")
    ax.set_ylabel("Kelompok Usia")
    ax.set_zlabel("Persentase (%)")
    ax.set_yticks([i * Y_SPACING for i in range(len(AGE_GROUPS))])
    ax.set_yticklabels(AGE_GROUPS)
    ax.set_title(detail_title)

    # ================= RAW TABLE =================
    table = "Jumlah Responden (N)\n"
    table += "Usia \\ Likert | " + " ".join(str(lv) for lv in LIKERT_VALUES) + "\n"
    for ag in AGE_GROUPS:
        table += f"{ag:<7} | " + " ".join(f"{raw[ag][lv]:>3}" for lv in LIKERT_VALUES) + "\n"

    plt.figtext(0.02, -0.25, table, ha="left", va="top",
                fontsize=8, family="monospace")

    safe = re.sub(r"[^A-Za-z0-9]+", "_", col)
    img_path = os.path.join(OUTPUT_DIR, f"{safe}_3D_norm.png")

    plt.savefig(img_path, dpi=150, bbox_inches="tight")
    plt.close()

    # ================= PPT SLIDE =================
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.title.text = group_title
    slide.shapes.add_picture(img_path, Inches(0.5), Inches(1.5), width=Inches(9))

# ================= SAVE =================
prs.save(PPT_PATH)
```

```python
def compute_group_scores(df, group_col, questions):
    
    tmp = df.copy()
    tmp[questions] = tmp[questions].apply(extract_likert)

    res = tmp.groupby(group_col)[questions].mean()

    res["mean_score"] = res.mean(axis=1)
    res["q25_score"] = res.quantile(0.25, axis=1)
    res["std_dev"] = res.std(axis=1)
    
    return (
    res[["mean_score","q25_score",'std_dev']]
    .sort_values('mean_score', ascending=False)
    
)
my_score = pd.DataFrame()
for g in [
    'Saya adalah seorang',
    'Age_Group',
    'Status pernikahan saya',
    'Pendidikan terakhir saya yang sudah saya selesaikan'
]:
    scores = compute_group_scores(data, g, myquestions)
    print("\n====", g, "====")
    print(scores)
    my_score=pd.concat([my_score, scores],axis=0)
```

```python
import pandas as pd


def compute_checkbox_counts_df(
    df,
    group_cols,
    checkbox_cols,
    percent=True
):
    """
    Return ONE dataframe containing all checkbox counts.

    Output columns:
    [group_cols..., question, option, count, percent]
    """

    all_results = []

    for col in checkbox_cols:

        tmp = df[group_cols + [col]].copy()

        # split answers
        tmp[col] = (
            tmp[col]
            .fillna('')
            .astype(str)
            .str.split(',')
        )

        # explode
        tmp = tmp.explode(col)

        # clean
        tmp[col] = tmp[col].str.strip()
        tmp = tmp[tmp[col] != '']

        # count
        counts = (
            tmp.groupby(group_cols + [col])
               .size()
               .reset_index(name='count')
        )

        # percent inside each group
        if percent:
            totals = counts.groupby(group_cols)['count'].transform('sum')
            counts['percent'] = counts['count'] / totals * 100

        # rename columns
        counts = counts.rename(columns={col: 'option'})
        counts['question'] = col

        all_results.append(counts)

    # combine everything
    final_df = pd.concat(all_results, ignore_index=True)

    return final_df
```

```python
group_cols = [
    'Saya adalah seorang',
    'Status pernikahan saya',
    'Pendidikan terakhir saya yang sudah saya selesaikan',
    'Age_Group'
]

checkbox_cols = [
    'Sebutkan hal-hal yang membuat anda merasa iman anda semakin bertumbuh.',
    'Kegiatan apa saja yang telah anda ikuti dalam Tahun Yubileum di Gereja Maria Bunda Karmel?'
]

checkbox_df = compute_checkbox_counts_df(others, group_cols, checkbox_cols)
```
